```python
# The number of passing students is represented as the height of a right-angled triangle.
# The number of rows corresponds to the number of passing students.
height = int(input("Enter the number of passing students (height of triangle): "))

# Outer loop to iterate through each row (from 1 up to 'height')
for row in range(1, height + 1):
    # Inner loop to print stars for the current 'row'
    # The number of stars is equal to the current row number
    for star in range(row):
        # Print a star, and use end="" to prevent a newline after each star
        print("*", end="")
    
    # Print a newline character after the inner loop finishes 
    # to move to the next row
    print()
```
