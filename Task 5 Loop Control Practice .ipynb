```python
def print_multiplication_table(n):
    """
    Prints the multiplication table from 1x1 up to nxn.
    """
    
    # 1. Print the header row (1 to n)
    # The 'end' argument prevents a newline, and f-string formats the output
    print(" x |", end="")
    for j in range(1, n + 1):
        print(f" {j:3}", end="") # :3 ensures 3 spaces for alignment
    print("\n" + "---" * (n + 1) + "-") # Print a separator line

    # 2. Outer Loop (Rows)
    for i in range(1, n + 1):
        # Print the row header (the current number i)
        print(f"{i:2} |", end="") 
        
        # 3. Inner Loop (Columns)
        for j in range(1, n + 1):
            product = i * j
            # Print the product, using :3 for neat column alignment
            print(f" {product:3}", end="")
            
        # 4. End of Inner Loop: Print a newline to move to the next row
        print()

# The number you provided
n = 5

print_multiplication_table(n)
```

```python
n = 5 

print(f"\n--- Modified Multiplication Table up to {n}x{n} ---\n")
print("Skipping cells where Row Number (i) = Column Number (j)\n")

# Outer loop (controls the row, i)
for i in range(1, n + 1):
    
    # Inner loop (controls the column, j)
    for j in range(1, n + 1):
        
        # Check the condition: if row number (i) equals column number (j)
        if i == j:
            # Use 'continue' to skip the rest of the inner loop body
            # for this specific iteration (i.e., skip calculating and printing)
            print("  -  ", end="") # Print a placeholder for clarity
            continue
            
        # If the condition is False, calculate and print the product
        product = i * j
        
        # Print the product, aligning it neatly
        print(f"{product:5}", end="") 
        
    # Print a new line after the inner loop finishes
    print()

print("\n----------------------------------")
```
