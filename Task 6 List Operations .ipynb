```python
# 1. Create an initial list of 5 cities
cities = ["Alexandria", "Paris", "Hawai", "London", 'Sydney']
print(f"Initial list: {cities}")
```

```python
# 2. Replace the 3rd city with a new city ('Monaco')
# The 3rd city is at index 2 (lists are 0-indexed)
cities[2] = 'Monaco'
print(f"After replacing 3rd city: {cities}")
```

```python
# 3. Append 2 more cities ('Rome' and 'Cairo')
cities.append('Rome')
cities.append('Cairo')
# Alternatively: cities.extend(['Rome', 'Cairo'])
print(f"After appending 2 more cities: {cities}")
```

```python
# 4. Remove the first city
# The first city is at index 0
cities.pop(0)
# Alternatively: del cities[0] or cities.remove('London') (if 'London' is guaranteed to be the first element)
print(f"After removing the first city: {cities}")
```

```python
# --- Print the final list and its length ---
# The final list
final_list = cities

# The length of the final list
list_length = len(final_list)

print("\n--- Final Results ---")
print(f"Final List: {final_list}")
print(f"Length of the Final List: {list_length}")
```
