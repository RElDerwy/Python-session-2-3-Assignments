```python
def print_multiplication_table(n):
    """
    Prints the multiplication table from 1x1 up to nxn.
    """
    
    # 1. Print the header row (1 to n)
    # The 'end' argument prevents a newline, and f-string formats the output
    print(" x |", end="")
    for j in range(1, n + 1):
        print(f" {j:3}", end="") # :3 ensures 3 spaces for alignment
    print("\n" + "---" * (n + 1) + "-") # Print a separator line

    # 2. Outer Loop (Rows)
    for i in range(1, n + 1):
        # Print the row header (the current number i)
        print(f"{i:2} |", end="") 
        
        # 3. Inner Loop (Columns)
        for j in range(1, n + 1):
            product = i * j
            # Print the product, using :3 for neat column alignment
            print(f" {product:3}", end="")
            
        # 4. End of Inner Loop: Print a newline to move to the next row
        print()

# The number you provided
n = 5

print_multiplication_table(n)
```
