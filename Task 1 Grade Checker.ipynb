```python
# list of sample student scores (out of 100)
scores = [95, 80, 45, 78, 55]

student_scores = [95, 80, 45, 78, 55]

for score in student_scores:
    # Outer grading structure (A, B, C, F)
    if score >= 90:
        grade = "A"
        # Nested check for "Excellent!"
        if score >= 85:
            print(f"Score: {score} -> Grade: {grade}, Excellent!")
        else:
            print(f"Score: {score} -> Grade: {grade}")
    elif score >= 75:
        grade = "B"
        # Nested check for "Excellent!"
        if score >= 85:
            print(f"Score: {score} -> Grade: {grade}, Excellent!")
        else:
            print(f"Score: {score} -> Grade: {grade}")
    elif score >= 50:
        grade = "C"
        # Nested check for "Excellent!" (Won't trigger here, but included for completeness of the logic placement)
        if score >= 85:
            print(f"Score: {score} -> Grade: {grade}, Excellent!")
        else:
            print(f"Score: {score} -> Grade: {grade}")
    else:
        grade = "F"
        print(f"Score: {score} -> Grade: {grade}")
```

```python
student_scores = [85, 98, 60, 74, 50]

for score in student_scores:
    # Determine the primary grade (A, B, C, F)
    if score >= 90:
        grade = "A"
    elif score >= 75:
        grade = "B"
    elif score >= 50:
        grade = "C"
    else:
        grade = "F"

    # Nested check for "Excellent!" (Applies to scores >= 85)
    # Note that the score 85 receives a grade of B (since it is less than 90 but greater than 75) AND it triggers the nested "Excellent!" message.
    if score >= 85:
        print(f"Score: {score} -> Grade: {grade}, Excellent!")
    else:
        # Only prints the grade if "Excellent!" isn't printed
        print(f"Score: {score} -> Grade: {grade}")
```

```python
def get_student_grade(score):
    """
    Calculates the letter grade and prints a message for a given score (0-100).
    A: >= 90
    B: >= 75
    C: >= 50
    F: < 50
    Prints 'Excellent!' if score is >= 85.
    """
    
    # Step 1: Determine the primary grade
    if score >= 90:
        grade = "A"
    elif score >= 75:
        grade = "B"
    elif score >= 50:
        grade = "C"
    else:
        grade = "F"

    # Step 2: Nested check for 'Excellent!' and print the result
    if score >= 85:
        print(f"Score: {score} -> Grade: {grade}, Excellent!")
    else:
        print(f"Score: {score} -> Grade: {grade}")

# --- Example Usage ---
new_scores = [92, 85, 70, 49, 100]

print("--- Function Output ---")
for s in new_scores:
    get_student_grade(s)
```
