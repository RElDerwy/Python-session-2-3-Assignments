```python
print("--- Even Numbers Between 1 and 50 ---")
# The range function goes up to, but does not include, the second number.
# So, we use range(1, 51) to include 50.
for number in range(1, 51):
    # Check if the number is even (remainder when divided by 2 is 0)
    if number % 2 == 0:
        print(number)

print("\n--- Odd Numbers Between 51 and 100 ---")
# We use range(51, 101) to include 100.
for number in range(51, 101):
    # Check if the number is odd (remainder when divided by 2 is NOT 0)
    if number % 2 != 0:
        print(number)
```

```python
# Even Numbers (Starts at 2, stops before 51, steps by 2)
print("Even Numbers (1-50):", list(range(2, 51, 2)))

# Odd Numbers (Starts at 51, stops before 101, steps by 2)
print("Odd Numbers (51-100):", list(range(51, 101, 2)))
```
